2a. Create a dataframe ```profiles``` containing the company profiles of all companies in the dow jones dataframe. Check the documentation for **profile** and use a loop to retrieve this data to store it in a dataframe. Include a check on whether or not the request was successful.

Some comments:
- first, create an empty dataframe to which you can append the information per iterated symbol
- follow the steps from exercise 1 inside the loop
- use the append method on the (initially) empty dataframe. **NOTE:** Unlike with lists, this method does not append in place. You need ```profiles = profiles.append...```

````python
from tqdm import tqdm
df_symbol=df.symbol.tolist()
filling='profile/'
profiles=pd.DataFrame()
for sym in tqdm(df_symbol):
    response=re.get(base_url+filling+sym,params=params)
    if response.status_code!=200:
        print ('ERROR !!!')
        break
    tempo_df=pd.DataFrame.from_dict(response.json())
    #tempo_df['symbol']=symbol
    profiles = pd.concat([profiles,tempo_df])
    del tempo_df

profiles
````
